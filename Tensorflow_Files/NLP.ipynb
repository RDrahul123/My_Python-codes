```python
text = "I LOVED this movie!!! üòçüòç It was worth Rs.500."
print(text)
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)  # keep only letters and spaces
    text = re.sub(r'\s+', ' ', text).strip()
    return text

cleaned_text = clean_text(text)
print(cleaned_text)
```

```python
from nltk.tokenize import word_tokenize
import nltk

nltk.download('punkt')
nltk.download('punkt_tab')

tokens = word_tokenize(cleaned_text)
print(tokens)
```

```python
from nltk.corpus import stopwords

nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

filtered_tokens = [word for word in tokens if word not in stop_words]
print(filtered_tokens)
```

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
print(stemmed_tokens)
```

```python
from nltk.stem import WordNetLemmatizer

nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()

lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]
print(lemmatized_tokens)
```

```python
final_text = " ".join(lemmatized_tokens)
print(final_text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "I love this movie",
    "This movie is terrible",
    "I love NLP",
    "NLP is amazing"
]

vectorizer = CountVectorizer()
bow_matrix = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(bow_matrix.toarray())
```

```python
from transformers import pipeline

# Load translation pipelines
en_to_fr = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")
fr_to_en = pipeline("translation_fr_to_en", model="Helsinki-NLP/opus-mt-fr-en")

def back_translate(sentence):
    # Step 1: English ‚Üí French
    french = en_to_fr(sentence)[0]["translation_text"]

    # Step 2: French ‚Üí English
    back_english = fr_to_en(french)[0]["translation_text"]

    return back_english

# Example
text = "Natural language processing is transforming modern AI."
augmented_text = back_translate(text)

print("Original:", text)
print("Back translated:", augmented_text)
```

```python
import random

def bigram_flip(sentence, flip_prob=0.3):
    words = sentence.split()
    result = words[:]

    for i in range(len(words) - 1):
        if random.random() < flip_prob:
            # Swap adjacent words
            result[i], result[i + 1] = result[i + 1], result[i]

    return " ".join(result)

# Example
text = "Natural language processing models learn patterns from data"
augmented_text = bigram_flip(text, flip_prob=0.4)

print("Original:", text)
print("Bigram flipped:", augmented_text)
```
